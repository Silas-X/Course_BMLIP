```julia
# Paramters
Σ=1e5*Diagonal(I,3) # I is the identity matrix of any dimension from package LinearAlgebra
σ2=2.0

# Generate data set
w=[1.0;2.0;0.25]
N=30
z=10.0*rand(N)
x_train=[[1.0;z;z^2] for z in z]
# function f(x)
#     (w'*x)[1]
# end
fig=plt.figure()
f(x)=(w'*x)[1]
y_train=map(f,x_train)+sqrt(σ2)*randn(N) # y[i]=w'*x[i]+ϵ
scatter(z,y_train);
xlabel(L"z");
ylabel(L"f([1.0,z,z^2])+ϵ");
# plt.show()
display(gcf())
```

```julia
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
f(x)=(w'*x)[1]
y_train=map(f,x_train)+sqrt(σ2)*randn(N) # y[i]=w'*x[i]+ϵ
ax.scatter(z,y_train);
ax.set_xlabel(L"z");
ax.set_ylabel(L"f([1.0,z,z^2])+ϵ");
fig.show()
fig
# display(gcf())
```
